The objectives and tasks of this notebook is based on an assignment in the [Applied Plotting, Charting & Data Representation in Python Course by University of Michigan on Coursera](https://www.coursera.org/learn/python-plotting?specialization=data-science-python).

The data correspont to 24 stations near **Ann Arbor, Michigan, United States**. The data comes from a subset of The National Centers for Environmental Information (NCEI) [Daily Global Historical Climatology Network](https://www.ncei.noaa.gov/access/search/data-search/daily-summaries)(GHCN-Daily). The GHCN-Daily is comprised of daily climate records from thousands of land surface stations across the globe.

Each row corresponds to a single day of observation for a particular station.

Inside the data files there are many variables, but for this case we only interested in the following:

* **STATION** : station identification code
* **DATE** : date in YYYY-MM-DD format (e.g. 2012-01-24 = January 24, 2012)
* **LATITUDE**: Latitude (decimal degrees)
* **LONGITUDE**: Longitude (decimal degrees)
* **NAME**: Name of the station
* **TMAX** : Maximum temperature (tenths of degrees C)
* **TMIN** : Minimum temperature (tenths of degrees C)

More information about the GHCN-Daily dataset [Here](https://www.ncei.noaa.gov/metadata/geoportal/rest/metadata/item/gov.noaa.ncdc:C00861/html)

Objective:

1. Overlay a scatter of the 2015 data for any points (highs and lows) for which the ten year record (2005-2014) record high or record low was broken in 2015.
2. Watch out for leap days (i.e. February 29th), it is reasonable to remove these points from the dataset for the purpose of this visualization.
3. Make the visual nice!

## Setting up the environment

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import glob
import datetime
```

```python
import cartopy.crs as ccrs
import cartopy.io.img_tiles as cimgt
```

## 1. Import Data

```python
# Relative path csv files
path = "../data/"

# Get all csv files in the csv folder
csv_files = glob.glob(os.path.join(path, "*.csv"))
```

```python
records = pd.DataFrame()

# Columns of interest
columns = ["STATION", "DATE", "LATITUDE", "LONGITUDE", "NAME", "TMAX", "TMIN"]

# Loop over all files in csv_files
for file in csv_files:
    # Temporaly dataframe
    temp = pd.read_csv(file, usecols=columns)
    
    # Join the temp and records dataframes
    records = pd.concat([records, temp])
```

```python
# Quick overview of the data
print("Rows:", records.shape[0])
print("Columns:", records.shape[1])
print('-' * 25)
print(records.dtypes)
print('-' * 25)
records.head()
```

## 2. Display Stations Location

```python
# Subset Columns
plot_data = records[["LONGITUDE", "LATITUDE", "NAME"]]

# Remove duplicated stations
plot_data = plot_data.drop_duplicates("NAME", ignore_index=True)

print(25 * '-')
print("Rows:", plot_data.shape[0])
print("Columns:", plot_data.shape[1])
print(25 * '-')

plot_data.head()
```

```python
lons = plot_data['LONGITUDE'].tolist()
lats = plot_data['LATITUDE'].tolist()
stations = plot_data['NAME'].tolist()

# Figure
fig = plt.figure(figsize=(10,10))

# Define and axes with a projection, ccrs.PlateCarree() allows us to work with latitude and longitude
ax = fig.add_subplot(1,1,1, projection=ccrs.PlateCarree())

# plot the station position
plt.scatter(plot_data['LONGITUDE'], plot_data['LATITUDE'], c='r', alpha=0.7, s=70)

# plot station name
offset = 0.006
for index, station in enumerate(stations):
    plt.text(lons[index] + offset, lats[index] + offset, station, fontsize=7.5)

# add basemap using the Open Street Map (OSM)
request = cimgt.OSM()
ax.add_image(request, 12)  # 12 is the zoom level
```

## 3. Prepare Data

### 3.1 Subset Data

```python
# Remove NAME, LATITUDE, and LONGITUDE columns
records = records[['STATION', 'DATE', 'TMAX', 'TMIN']]
```

### 3.2 Convert DATE to datetime

```python
records['DATE'] = pd.to_datetime(records['DATE'])
```

### 3.3 Filter Data

```python
# Filter out all observations prior 2005-01-01 and after 2015-12-31
records = records[records['DATE'] >= '2005-01-01']
records = records[records['DATE'] < '2016-01-01']

# Check result
print("Min. day on records: " , records['DATE'].min())
print("Max. day on records: ", records['DATE'].max())

# mask feb-29
mask = (records['DATE'] == '2008-02-29') | (records['DATE'] == '2012-02-29')

# Mask out the dates 29 of february 2008 and 2012
records = records[~mask]
```

### 3.4 Convert data format to long

```python
records = records.melt(id_vars=['STATION', 'DATE'], value_vars=['TMAX', 'TMIN'], var_name="temperature")
records.rename(columns={'STATION':'id', 'DATE': 'date'}, inplace=True)

# display result
records.head()
```

### 3.4 Convert tenths of degree to degree

```python
records['value'] = records['value'].apply(lambda x: x / 10)
```

### 3.5 Compute minimum and maximum daily temperature

```python
# There are 24 observations (stations) per day, we select the lowest and highest temperature value.
records = records.groupby('date').agg({'value': ['min', 'max']})

# Remove the extra column level
records = records.droplevel(level=0, axis=1)

# Reset index
records.reset_index(inplace=True)

# rename min and max column
records.rename(columns={'min': 'tmin','max': 'tmax'}, inplace=True)

# Display result
records.head()
```

### 3.6 Add a column for the day of the year

```python
# Define a function that return the day of the year of a given date. The function in leap years (2008 and 2012)
# correct every date after the 29 of february
def get_day_of_year(date):
    day_of_year = pd.Period(date, freq='D').dayofyear
    
    if ((date.year == 2012) | (date.year == 2008)) & (date.month > 2):
        day_of_year -= 1
            
    return day_of_year
```

```python
records['day_of_year'] = records['date'].apply(lambda x: get_day_of_year(x))
records.head()
```

### 3.7 Split dataframe

```python
mask = records['date'] >= '2015-01-01'

# Get only the records for 2015, and rest the index numeration
records_2015 = records[mask].set_index('day_of_year')

# remove the records for 2015
records = records[~mask]
```

```python
# Overview of records_2015
print('Rows: ', records_2015.shape[0])
print('columns: ', records_2015.shape[1])
print('-' * 40)
print(records_2015.head())
```

### 3.8 Lowest and highest temperature recorded by day of year between 2005 and 2014

```python
records = records.groupby('day_of_year').agg({'tmin': 'min', 'tmax': 'max'})
```

```python
# Overview of records
print('Rows: ', records.shape[0])
print('columns: ', records.shape[1])
print('-' * 40)
print(records.head())
```

### 3.9 Temperature record in 2015, compared with 2005-2014 period

```python
# Get the dates where was a record of minimum or maximum temperature in 2015
data_2015_min = records_2015[records_2015['tmin'] < records['tmin']]['tmin']
data_2015_max = records_2015[records_2015['tmax'] > records['tmax']]['tmax']
```

## 4. Plot

```python
plt.figure(figsize=(10,10) ,facecolor='white')

# Create the line plots for the data between 2005-2015
plt.plot(records.index, records['tmin'], c='#74a8aa',linewidth=0.6)
plt.plot(records.index, records['tmax'], c='#fdb353', linewidth=0.6)

# Create the scatter plot for the record temperature found in 2015
plt.scatter(data_2015_min.index, data_2015_min.values, s=2.5, c='black')
plt.scatter(data_2015_max.index, data_2015_max.values, s=2.5, c='black')

# Set title, labels and legend
plt.title('Record temperatures registered in 2015, compared to observations\nmade between 2005 and 2014 using data from 24 stations\nin Michigan, USA.',
          loc='left',  fontsize=18)
plt.ylabel('Temperature (°C)', alpha=0.8)
plt.xlabel('Day of the Year', alpha=0.8)
plt.legend(['Minimum between 2005-2014', 'Maximum between 2005-2014', 'Record in 2015'], title='Temperature',
          fontsize=8)

# fill the area between the linear data and exponential data (lecture code)
plt.gca().fill_between(range(365), 
                       records['tmin'], records['tmax'], 
                       facecolor='gray', 
                       alpha=0.08)

# remove all the ticks (both axes), and tick labels on the Y axis
plt.tick_params(top=False, bottom=False, left=False, right=False, labelleft=True, labelbottom=True)

# adjust the top of the plot 
plt.subplots_adjust(top=0.80)

# remove the top and right frame of the chart
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.show()
```
